### VERİ GÖRSELLEŞTİRME

```python
df["neighbourhood_group"].value_counts().plot.barh().set_title("NEIGHBOURHOOD GROUP")
```

```python
sns.countplot(data=df,x="neighbourhood_group", palette="gray")
```

```python
sns.barplot(x="neighbourhood_group" , y="price", data=df)
plt.title("NEIGHBOURHOOD GROUP PRICE")
```

```python
df.groupby("neighbourhood_group").sum().plot.pie(y = "price", autopct="%.1f%%", ylabel="", legend=False, figsize=(5,10))
```

```python
sns.countplot(data=df, x="room_type", palette="pink")
plt.title("ROOM TYPES")
```

```python
sns.barplot(x="room_type", y="price", data=df , palette="pink")
plt.title("ROOM TYPES AND RATES ")
```

```python
room_type_total_earnings = df.groupby("room_type")["price"].sum()

plt.figure(figsize=(8, 5))
plt.pie(room_type_total_earnings, labels=room_type_total_earnings.index, autopct='%1.1f%%', startangle=150)
plt.title("TOTAL REVENUE DISTRIBUTION BY ROOM TYPES")
```

```python
top_expensive_listings = df.nlargest(11, 'price')[['neighbourhood', 'name', 'price']]
plt.figure(figsize=(12, 8))
sns.barplot(x='neighbourhood', y='price', data=top_expensive_listings, )
plt.title("TOP 10 MOST EXPENSİVE NEİGHBOURHOODS")
plt.xlabel('Neighbourhood')
plt.ylabel('Price')
plt.xticks(rotation=35)
```

```python
top_reviewed_neighbourhoods = df["neighbourhood"].value_counts().nlargest(6).index
df_top_reviews = df[df["neighbourhood"].isin(top_reviewed_neighbourhoods)]
average_prices_by_neighbourhood = df_top_reviews.groupby("neighbourhood")["price"].mean().reset_index()

plt.figure(figsize=(12, 9))
sns.barplot(x="neighbourhood" , y="price", data=average_prices_by_neighbourhood)
plt.title("AVERAGE PRICES IN NEIGHBOURHOODS WITH THE MOST COMMENTS")
plt.xlabel("Neighbourhood")
plt.ylabel("Average Price")
plt.xticks(rotation=35)
```

```python
sns.catplot(x="neighbourhood_group", y="number_of_reviews", data=df, hue="room_type", palette="rocket")
```

```python
top_reviewed_neighbourhoods = df['neighbourhood'].value_counts().nlargest(6).index
df_top_reviews = df[df ["neighbourhood"].isin(top_reviewed_neighbourhoods)]
total_reviews_by_neighbourhood = df_top_reviews["neighbourhood"].value_counts()

plt.figure(figsize=(12, 8 ))
sns.histplot(data=df_top_reviews, x="neighbourhood", bins=len(top_reviewed_neighbourhoods), kde=True)
plt.title("TOP 6 NEIGHBOURHOODS WITH THE MOST REVIEWS")
plt.xlabel("Neighbourhood")
plt.ylabel("Total Number Of Reviews")
plt.xticks(rotation=35)
```
